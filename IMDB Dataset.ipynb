### Histograms
The runtime (mins) is right skewed, clusterd around 270 min mark. This is telling us that more movies have longer run times

```python
var = df["Rating"].var()
print(var)
```

```python
std = df["Rating"].std()
print(std)
```

```python
top_10_ratings = df.nlargest(n=10, columns=['Rating'])
print(top_10_ratings)
```

```python
bottom_5_ratings = df.nsmallest(n=5, columns=['Rating'])
print(bottom_5_ratings)
```

```python
above_85 = df[df['Rating'] > 8.5]
print(above_85)
```

```python
vote_correlation = df['Rating'].corr(df['Votes'])
print('Vote Correlation:', vote_correlation)
runtime_correlation = df['Rating'].corr(df['Runtime (Minutes)'])
print('Runtime Correlation:', runtime_correlation)
```

### Correlation
Voting is more correlated to the rating than runtime is. They both have positive correlations though.

```python
revenue_correlation = df['Rating'].corr(df['Revenue (Millions)'])
print('Revenue Correlation:', revenue_correlation)
metascore_correlation = df['Rating'].corr(df['Metascore'])
print('Metascore Correlation:', metascore_correlation)
rank_correlation = df['Rating'].corr(df['Rank'])
print('Rank Correlation:', rank_correlation)
```

```python
above8_and_less100 = df[(df['Rating'] > 8) & (df['Runtime (Minutes)'] < 100)]
print(above8_and_less100)
```
